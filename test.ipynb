```python
import pandas as pd
```

```python
def get_data(dir= 'data', split= 'train', track= 'a', language= 'ptbr'):
    
    archive = language + '.csv' if split == 'train' else language + '_' + track + '.csv'

    path = f'{dir}/{split}/track_{track}/{archive}'
    
    return pd.read_csv(path)
```

```python
df = get_data()
```

```python
df
```

```python
df2 = get_data(track= 'b')
```

```python
if set(df['text']) == set(df2['text']):
    print("As colunas são iguais, independentemente da ordem.")
else:
    print("As colunas são diferentes.")
```

```python
count = df['text'].isin(df2['text']).sum()

print(f"Número de valores da col1 presentes na col2: {count}")
```
